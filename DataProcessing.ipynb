```python
import os
os.chdir("C:\\Users\\khali\\OneDrive\\√Årea de Trabalho\\LSTM-FCN-master")

import librosa as lr
import pandas as pd
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
import random
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
data = glob('_data/all_data' + '/*.wav')
#random.shuffle(data)

label = []
for d in data:
    if 'H' in d:
        label.append('0')
    elif 'S' in d:
        label.append('1')
    elif 'happy' in d:
        label.append('0')
    elif 'angry' in d:
        label.append('1')
```

```python
labels = pd.DataFrame(label)
labels.columns = [-1]

features_data = pd.DataFrame(columns=['feature'])
bookmark=0
```

```python
new_feature = []
for file,y in enumerate(data):
    X,sample_rate = lr.load(data[file],res_type='kaiser_fast',duration=3,sr=8000*2,offset=0)
    sample_rate = np.array(sample_rate)
    mfccs = lr.feature.mfcc(y=X,sr=sample_rate,n_mfcc=15)
    new_feature.append(np.mean(mfccs[0]))
    new_feature.append(np.mean(mfccs[1]))
    new_feature.append(np.mean(mfccs[2]))
    new_feature.append(np.mean(mfccs[3]))
    new_feature.append(np.mean(mfccs[4]))
    new_feature.append(np.mean(mfccs[5]))
    new_feature.append(np.mean(mfccs[6]))
    new_feature.append(np.mean(mfccs[7]))
    new_feature.append(np.mean(mfccs[8]))
    new_feature.append(np.mean(mfccs[9]))
    new_feature.append(np.mean(mfccs[10]))
    new_feature.append(np.mean(mfccs[11]))
    new_feature.append(np.mean(mfccs[12]))
    new_feature.append(np.mean(mfccs[13]))
    new_feature.append(np.mean(mfccs[14]))
    
    features_data.loc[bookmark] = [new_feature]
    
    new_feature = []
    bookmark=bookmark+1
```

```python
df3 = pd.DataFrame(features_data['feature'].values.tolist())
```

```python
df3.tail()
```

```python
newdf = pd.concat([df3,labels], axis=1)
final_data = newdf.rename(index=str, columns={"0": "label"})
```

```python
final_data.fillna(0, inplace=True)
```
